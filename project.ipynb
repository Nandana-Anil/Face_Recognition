```python
import cv2
import numpy as np
import os
import face_recognition as fr
```

```python
path='images'
```

```python
mylist=os.listdir(path)
mylist
```

```python
#set input & output

imgs=[] #inputs
classnames=[] #outputs

for i in mylist:
    imgpath=os.path.join(path,i)
    current_img=cv2.imread(imgpath)
    imgs.append(current_img) #images to input list
    classnames.append(i.split('.')[0]) #set output names
    # print(classnames)
```

```python
#pre-processing

#BGR (since imread) -> RGB

#need - face_location, face_encoding

img=imgs[0]
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #convert to rgb
face_in_frame=fr.face_locations(img) #gets face coordinates for face location
print(face_in_frame)
encoding=fr.face_encodings(img,face_in_frame)
encoding
```

```python
def faceencodings(images):
    encodelist=[]
    for i in images:
        i=cv2.cvtColor(i,cv2.COLOR_BGR2RGB)
        face_in_frames=fr.face_locations(i)
        # print(face_in_frames)
        encode=fr.face_encodings(i,face_in_frames)[0] #for 1 face from each image
        encodelist.append(encode)
    return encodelist
```

```python
encodelist_knownfaces=faceencodings(imgs)
encodelist_knownfaces
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,image2=video.read()
    image3=cv2.cvtColor(image2,cv2.COLOR_BGR2RGB)
    fif=fr.face_locations(image3)
    # print(fif)
    enc_face=fr.face_encodings(image3,fif)
    # print(enc_face)
    for face_loc,enc in zip(fif,enc_face):
        matches=fr.compare_faces(encodelist_knownfaces,enc)
        # print(matches)
        face_dist=fr.face_distance(encodelist_knownfaces,enc)
        # print(face_dist)

        matchIndex=np.argmin(face_dist)
        # print(matchIndex)
        if matches[matchIndex]==True:
            name=classnames[matchIndex]
            # print(name)

            y1,x2,y2,x1=face_loc
            # print(x1,y1)
            cv2.rectangle(image2,(x1,y1),(x2,y2),(0,255,255),2)
            cv2.putText(image2,name,(x1,y1),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),1)

    cv2.imshow('Video',image2)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
